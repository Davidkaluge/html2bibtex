````python

# ==========================================================================================
# MASTERPIECE SOFTWARE FACTORY - CELL 1: ORCHESTRATOR & DEVOPS INFRASTRUCTURE INITIALIZATION (V1.1)
# ==========================================================================================

# 1. Menginstal SDK Google GenAI Terbaru Secara Silent
!pip install -q -U google-genai

import os
import sys
import json
import subprocess
from google.colab import userdata

print("========================================================================")
print("🏢 INDUSTRIAL MULTI-AGENT SOFTWARE FACTORY - INITIALIZATION SYSTEM")
print("========================================================================\n")

# 2. INTEGRASI OTAK UTAMA (GEMINI API VIA COLAB SECRETS)
client = None
try:
    # Memanggil library secara dinamis setelah instalasi silent
    from google import genai
    from google.genai import types

    api_key_env = userdata.get('GEMINI_API_KEY')
    client = genai.Client(api_key=api_key_env)
    print("✅ [SYSTEM] Otak Utama (Gemini API) Berhasil Terhubung ke Pabrik.")
except userdata.SecretNotFoundError:
    print("❌ [CRITICAL ERROR] Kunci rahasia 'GEMINI_API_KEY' tidak ditemukan!")
    print("   👉 SOLUSI: Klik ikon KUNCI (Secrets) di panel kiri Colab.")
    print("   👉 Tambahkan nama: GEMINI_API_KEY")
    print("   👉 Masukkan Value: [API Key dari AI Studio Anda]")
    print("   👉 AKTIFKAN saklar 'Notebook Access' (geser sampai biru).")
except Exception as e:
    print(f"❌ [SYSTEM ERROR] Gagal inisialisasi: {e}")

# 3. IMPLEMENTASI AGENT DEVOPS (INFRASTRUCTURE GUARD)
class AgentDevOps:
    def __init__(self):
        self.name = "Agent DevOps (Infrastructure Guard)"

    def audit_environment(self):
        print(f"\n🔍 [AUDIT] {self.name} sedang memindai spesifikasi Google Colab...")
        python_version = sys.version.split()[0]
        print(f"   📊 OS Dasar      : Linux (Ubuntu Colab Environment)")
        print(f"   📊 Versi Python  : {python_version}")
        return python_version

    def safe_install_dependencies(self, required_packages):
        print(f"\n⚡ [DEVOPS] Memulai Resolusi Dependensi & Urutan Instalasi Sekuensial...")

        for package in required_packages:
            print(f"   📥 Menginstal paket steril: {package}...")
            try:
                subprocess.run(
                    [sys.executable, "-m", "pip", "install", "-q", package],
                    check=True,
                    stdout=subprocess.DEVNULL,
                    stderr=subprocess.DEVNULL
                )
                print(f"   ✅ {package} berhasil terpasang tanpa konflik.")
            except subprocess.CalledProcessError:
                if "==" in package:
                    fallback_package = package.split("==")[0]
                    print(f"   🔄 Mencoba memasang versi alternatif aman untuk: {fallback_package}...")
                    subprocess.run(
                        [sys.executable, "-m", "pip", "install", "-q", fallback_package],
                        stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL
                    )

        print("\n🧪 [SMOKE TEST] Menjalankan Uji Impor Kesiapan Infrastruktur...")
        try:
            import cryptography
            import pydantic
            print("   ✅ Uji Impor Sempurna: Seluruh library terisolasi dengan stabil.")
            return True
        except ImportError as e:
            print(f"   ❌ Uji Kesiapan Gagal: {e}. Terjadi tabrakan versi pustaka.")
            return False

# 4. IMPLEMENTASI AGENT ORCHESTRATOR (CEO & CHIEF ARCHITECT)
class AgentOrchestrator:
    def __init__(self, client_ai):
        self.client = client_ai
        self.name = "Agent Orchestrator (CEO)"
        self.master_task_list = {}
        self.devops = AgentDevOps()

    def initiate_factory(self, user_request, project_mode, distribution_type):
        if self.client is None:
            print("\n❌ [STOP] Pabrik tidak dapat berjalan karena API Key belum dikonfigurasi dengan benar.")
            return

        print(f"\n🎯 [ORCHESTRATOR] Menerima Perintah Proyek Baru: '{user_request}'")
        print(f"   🔹 Mode Bisnis       : {project_mode}")
        print(f"   🔹 Jenis Distribusi  : {distribution_type}")

        self.devops.audit_environment()
        dependencies_awal = ["pydantic==2.5.3", "cryptography", "requests"]
        infra_ready = self.devops.safe_install_dependencies(dependencies_awal)

        if not infra_ready:
            print("❌ [STOP] Pabrik dihentikan sementara karena infrastruktur tidak stabil.")
            return

        print(f"\n🚀 [SUKSES] {self.name}: Infrastruktur 100% SIAP dan STERIL.")
        print("📥 Menunggu untuk mengeksekusi CELL 2 (Pembuatan Master Task List & Pemanggilan Tim Agen)...")

# ==========================================================================================
# AREA EKSEKUSI UTAMA
# ==========================================================================================
if client is not None:
    orchestrator = AgentOrchestrator(client_ai=client)

    perintah_user = "Aplikasi Scanner Dokumen berbasis AI dengan sinkronisasi Cloud"
    mode_bisnis = "BISNIS (SaaS dengan Payment Gateway & Sistem Expired)"
    tipe_distribusi = "PARAMETER-DRIVEN (White-Label Ready)"

    orchestrator.initiate_factory(
        user_request=perintah_user,
        project_mode=mode_bisnis,
        distribution_type=tipe_distribusi
    )
````

````python
# ==========================================================================================
# MASTERPIECE SOFTWARE FACTORY - CELL 2: DYNAMIC AGENT CREATION & MASTER TASK LIST REGISTRY
# ==========================================================================================

print("========================================================================")
print("🧠 PHASE 2: ORCHESTRATION & CENTRALIZED DEPENDENCY REGISTRY ACTIVATION")
print("========================================================================\n")

# Prompt Instruksi Tingkat Tinggi untuk Mengatur Cara Berpikir Otak Orchestrator
PROMPT_ANALISIS_ORCHESTRATOR = """
Anda adalah Senior Enterprise Software Architect dan CEO AI Software Factory.
Tugas Anda adalah menganalisis permintaan proyek dari user dan menentukan:
1. Agen Teknis apa saja yang dibutuhkan secara dinamis (Sebutkan nama agen).
2. Tupoksi dan Daftar Tugas spesifik untuk masing-masing agen tersebut dalam format JSON terstruktur.

Aturan Penting:
- Jumlah agen harus fleksibel mengikuti kebutuhan proyek. Karena proyek ini melibatkan PC, HP, Cloud, Payment Gateway, dan Lisensi Kedaluwarsa, pastikan Anda memanggil agen spesialis yang tepat (Misal: Agen_Cloud_API, Agen_Desktop_Windows, Agen_Mobile_Android, Agen_FinTech_Security).
- Output HARUS murni berupa kode JSON valid, tanpa teks penjelasan tambahan di luar JSON, tanpa tanda ```json.

Format JSON yang wajib Anda ikuti:
{
  "nama_proyek": "Nama Proyek",
  "analisis_makro": "Deskripsi singkat arsitektur ekosistem",
  "tim_agen": {
    "Nama_Agent_A": {
      "tupoksi": "Tanggung jawab utama agen ini",
      "daftar_tugas": [
        "Tugas 1",
        "Tugas 2",
        "Tugas 3"
      ]
    },
    "Nama_Agent_B": {
      "tupoksi": "Tanggung jawab utama agen ini",
      "daftar_tugas": [
        "Tugas 1",
        "Tugas 2"
      ]
    }
  }
}
"""

def jalankan_fase_dua(orchestrator_instance, request_user, mode_bisnis, tipe_distribusi):
    print("🤖 Agent Orchestrator sedang membedah proyek di pikiran AI...")

    prompt_konten = f"""
    MOHON ANALISIS PROYEK BERIKUT:
    - Permintaan User: {request_user}
    - Model Bisnis: {mode_bisnis}
    - Tipe Distribusi Kode: {tipe_distribusi}

    Hasilkan Master Task List sesuai instruksi arsitektur!
    """

    try:
        # Memanggil Gemini API menggunakan model Flash untuk analisis cepat berstruktur
        response = orchestrator_instance.client.models.generate_content(
            model='gemini-2.5-flash',
            contents=prompt_konten,
            config=types.GenerateContentConfig(
                system_instruction=PROMPT_ANALISIS_ORCHESTRATOR,
                temperature=0.2 # Nilai rendah agar AI konsisten menghasilkan JSON murni
            )
        )

        # Membersihkan teks respons dari potensi sisa markdown AI
        clean_json_text = response.text.strip().replace("```json", "").replace("```", "")

        # Menyalin data ke Master Task List Pusat di dalam sistem Orchestrator
        orchestrator_instance.master_task_list = json.loads(clean_json_text)

        # Menampilkan Hasil Cetak Biru Pembagian Tugas Pusat
        print("\n📊 ==================================================================")
        print(f"📋 MASTER TASK LIST PUSAT: {orchestrator_instance.master_task_list.get('nama_proyek')}")
        print("====================================================================")
        print(f"💡 Analisis Makro Arsitektur:\n{orchestrator_instance.master_task_list.get('analisis_makro')}\n")

        print("👥 TIM AGEN AKTIF & TUPOKSI YANG DISINKRONISASIKAN:")
        print("--------------------------------------------------------------------")

        for nama_agent, detail in orchestrator_instance.master_task_list.get('tim_agen', {}).items():
            print(f"\n👨‍💻 NAMA AGEN: {nama_agent.upper()}")
            print(f"   🎯 Tupoksi: {detail.get('tupoksi')}")
            print("   📝 Daftar Tugas Terdaftar:")
            for index, tugas in enumerate(detail.get('daftar_tugas', []), 1):
                print(f"      {index}. {tugas}")

        print("\n====================================================================")
        print("✅ [SUKSES] Master Task List berhasil disalin ke Papan Informasi Pusat!")
        print("🚀 Hubungan antar-agen terkunci. Sistem siap masuk ke CELL 3 (Penulisan Kode Fisik).")

    except Exception as e:
        print(f"❌ [ERROR FASE 2] Gagal menyusun Master Task List: {e}")
        print("   Respons AI Mentah yang gagal di-parsing ke JSON:")
        print(response.text if 'response' in locals() else "Tidak ada respons")

# JALANKAN PROSES FASE 2
# (Menggunakan instance 'orchestrator' yang sudah sukses dibuat di Cell 1)
jalankan_fase_dua(
    orchestrator_instance=orchestrator,
    request_user=perintah_user,
    mode_bisnis=mode_bisnis,
    tipe_distribusi=tipe_distribusi
)
````

````python
# ==========================================================================================
# MASTERPIECE SOFTWARE FACTORY - CELL 3: AUTONOMOUS CODE PRODUCTION WITH AUTO-RETRY (V1.2)
# ==========================================================================================

print("========================================================================")
print("💻 PHASE 3: AUTOMATED CODE PRODUCTION & PHYSICAL FILE WRITING ACTIVATION")
print("========================================================================\n")

import time

# Prompt Instruksi Ketat untuk Agen Pemrogram Teknis
PROMPT_AGEN_PROGRAMMER = """
Anda adalah Senior Full-Stack Engineer spesifik untuk agen yang ditunjuk.
Tugas Anda adalah menulis KODE UTUH, BERSIH, dan SIAP PAKAI untuk file yang diminta berdasarkan Master Task List proyek.

Aturan Penulisan Kode:
1. Patuhi standar penulisan kode internasional (PEP 8 untuk Python, Clean Code Architecture).
2. Terapkan prinsip PARAMETER-DRIVEN: Jangan menanam API Key atau URL Server langsung di kode. Baca nilai-nilai tersebut dari file konfigurasi eksternal (seperti `config.json` atau menggunakan environment variables).
3. Terapkan sistem AAA: Kode harus mendukung registrasi, login, dan pengecekan masa aktif (expired date) yang terhubung ke cloud backend.
4. Tangani error secara ramah (Graceful Error Handling): Gunakan blok try-except agar aplikasi tidak force close jika terjadi kendala jaringan.
5. Sampaikan output MURNI berupa kode pemrograman di dalam blok markdown (contoh: ```python ... ```). Jangan berikan teks penjelasan teoretis, langsung berikan kode siap pakai.
"""

def jalankan_fase_tiga_v2(orchestrator_instance):
    if not orchestrator_instance.master_task_list:
        print("❌ [STOP] Master Task List kosong. Jalankan CELL 2 terlebih dahulu!")
        return

    root_output_dir = "Software_Factory_Output"
    os.makedirs(root_output_dir, exist_ok=True)
    print(f"📁 [DISK] Membuat/memastikan direktori utama: '{root_output_dir}/'")

    tim_agen = orchestrator_instance.master_task_list.get("tim_agen", {})
    nama_proyek = orchestrator_instance.master_task_list.get("nama_proyek", "Aplikasi Enterprise")

    for nama_agen, detail in tim_agen.items():
        print(f"\n📢 [ORCHESTRATOR] Memanggil {nama_agen.upper()} untuk memproduksi modul...")

        sub_dir = os.path.join(root_output_dir, nama_agen.lower())
        os.makedirs(sub_dir, exist_ok=True)

        tupoksi = detail.get("tupoksi")
        daftar_tugas = detail.get("daftar_tugas", [])
        nama_file_target = f"{nama_agen.lower()}_core.py"

        prompt_konten_kode = f"""
        PROYEK: {nama_proyek}
        PERAN ANDA: {nama_agen}
        TUPOKSI ANDA: {tupoksi}
        DAFTAR TUGAS ANDA: {json.dumps(daftar_tugas)}

        Tuliskan kode Python utama bernama '{nama_file_target}' secara utuh, menerapkan seluruh aturan Parameter-Driven, AAA System, dan Graceful Error Handling.
        """

        # SISTEM AUTO-RETRY KETAT (MAX RETRIES = 3) UNTUK MENGATASI ERROR 503
        max_percobaan = 3
        sukses_produksi = False

        for percobaan in range(1, max_percobaan + 1):
            try:
                response_kode = orchestrator_instance.client.models.generate_content(
                    model='gemini-2.5-flash',
                    contents=prompt_konten_kode,
                    config=types.GenerateContentConfig(
                        system_instruction=PROMPT_AGEN_PROGRAMMER,
                        temperature=0.3
                    )
                )

                # Proses pembersihan kode jika pemanggilan berhasil
                raw_code = response_kode.text.strip()
                clean_code = raw_code
                if "```python" in raw_code:
                    clean_code = raw_code.split("```python")[1].split("```")[0]
                elif "```" in raw_code:
                    clean_code = raw_code.split("```")[1].split("```")[0]

                path_file_final = os.path.join(sub_dir, nama_file_target)
                with open(path_file_final, "w") as berkas_file:
                    berkas_file.write(clean_code.strip())

                print(f"   💾 [DISK SUCCESS] File '{nama_file_target}' berhasil ditulis di '{sub_dir}/' (Percobaan ke-{percobaan})")
                sukses_produksi = True
                break # Keluar dari loop retry karena sudah sukses

            except Exception as e:
                print(f"   ⚠️ [PERCOBAAN {percobaan}/{max_percobaan} GAGAL] Server Google sibuk (Error 503).")
                if percobaan < max_percobaan:
                    jeda_waktu = percobaan * 5 # Jeda waktu bertambah lama (5s, 10s) memberi nafas pada server
                    print(f"   🔄 Mengambil nafas... Menunggu {jeda_waktu} detik sebelum mencoba kembali...")
                    time.sleep(jeda_waktu)
                else:
                    print(f"   ❌ [CRITICAL] {nama_agen} gagal berproduksi setelah {max_percobaan} kali mencoba akibat lonjakan server.")

        print("--------------------------------------------------------------------")

    # Pembuatan file config.json eksternal
    config_template = {
        "APP_NAME": nama_proyek,
        "VERSION": "1.0.0",
        "CLOUD_SERVER_URL": "https://api.pabriksoftware.cloud/v1",
        "PAYMENT_GATEWAY_SANDBOX_KEY": "SB-Mid-Xxxxx-Xxxxx",
        "ENCRYPTION_ALGORITHM": "AES-256",
        "LICENSE_CHECK_INTERVAL_DAYS": 1
    }

    with open(os.path.join(root_output_dir, "config.json"), "w") as config_file:
        json.dump(config_template, config_file, indent=4)
    print("\n⚙️ [CONFIG] File 'config.json' terenkapsulasi berhasil dibuat di folder utama.")

    print("\n====================================================================")
    print("🚀 [PRODUKSI ITERASI SELESAI] Silakan cek kembali folder log di kiri Colab Anda.")
    print("====================================================================")

# Jalankan Ulang Fase 3 dengan Sistem Proteksi Baru
jalankan_fase_tiga_v2(orchestrator_instance=orchestrator)
````

````python
# ==========================================================================================
# MASTERPIECE SOFTWARE FACTORY - CELL 3: BULK PRODUCTION ARCHITECTURE (ANTI-503 LIMIT)
# ==========================================================================================

print("========================================================================")
print("💻 PHASE 3: BULK CODE PRODUCTION & AUTOMATED INJECTION (ANTI-RATE-LIMIT)")
print("========================================================================\n")

import time

def jalankan_fase_tiga_bulk(orchestrator_instance):
    if not orchestrator_instance.master_task_list:
        print("❌ [STOP] Master Task List kosong. Jalankan CELL 2 terlebih dahulu!")
        return

    root_output_dir = "Software_Factory_Output"
    os.makedirs(root_output_dir, exist_ok=True)

    tim_agen = orchestrator_instance.master_task_list.get("tim_agen", {})
    nama_proyek = orchestrator_instance.master_task_list.get("nama_proyek", "Aplikasi Enterprise")

    # Menyusun instruksi massal untuk menghemat pemanggilan API
    deskripsi_tim = ""
    for nama_agen, detail in tim_agen.items():
        deskripsi_tim += f"- AGEN: {nama_agen}\n  Tupoksi: {detail.get('tupoksi')}\n  Tugas: {json.dumps(detail.get('daftar_tugas'))}\n\n"

    PROMPT_BULK = f"""
    Anda adalah Chief Full-Stack Engineer. Tugas Anda adalah memproduksi KODE UTUH untuk seluruh tim agen sekaligus.

    PROYEK: {nama_proyek}
    DAFTAR TIM & TUGAS NYA:
    {deskripsi_tim}

    ATURAN KODE:
    1. Terapkan PARAMETER-DRIVEN (baca dari `config.json` atau env).
    2. Terapkan AAA System (Autentikasi, Login, Cek Expired Date dari Cloud secara real-time).
    3. Terapkan Graceful Error Handling (try-except agar anti-crash).

    Format Output HARUS berupa satu objek JSON bersih (tanpa markdown, tanpa teks penjelasan) dengan struktur berikut:
    {{
       "NAMA_AGEN_A": "isi kode python utuh untuk agen A",
       "NAMA_AGEN_B": "isi kode python utuh untuk agen B"
    }}
    """

    print("🤖 [ORCHESTRATOR] Mengirim instruksi Bulk Production ke pusat kendali AI...")

    max_retries = 3
    response_text = ""
    for attempt in range(1, max_retries + 1):
        try:
            response = orchestrator_instance.client.models.generate_content(
                model='gemini-2.5-flash',
                contents=PROMPT_BULK,
                config=types.GenerateContentConfig(
                    temperature=0.2
                )
            )
            response_text = response.text.strip().replace("```json", "").replace("```", "")
            break
        except Exception as e:
            print(f"   ⚠️ [PERCOBAAN {attempt}] Server masih sibuk. Menunggu 7 detik...")
            time.sleep(7)

    if not response_text:
        print("❌ [CRITICAL] Gagal total menghubungi server. Mohon tunggu beberapa saat lagi.")
        return

    try:
        # Membongkar paket kode massal dari AI
        bundel_kode = json.loads(response_text)

        # Menyebarkan kode ke sub-folder fisik masing-masing agen
        for nama_agen, detail in tim_agen.items():
            sub_dir = os.path.join(root_output_dir, nama_agen.lower())
            os.makedirs(sub_dir, exist_ok=True)

            nama_file_target = f"{nama_agen.lower()}_core.py"
            path_file_final = os.path.join(sub_dir, nama_file_target)

            # Ambil kode spesifik untuk agen ini dari bundel JSON
            isi_kode = bundel_kode.get(nama_agen, f"# Kode otomatis untuk {nama_agen}\nprint('Module Active')")

            with open(path_file_final, "w") as berkas_file:
                berkas_file.write(isi_kode.strip())

            print(f"💾 [DISK SUCCESS] Berkas '{nama_file_target}' berhasil di-inject ke '{sub_dir}/'")

    except Exception as parse_error:
        print(f"⚠️ Gagal ekstraksi otomatis, mengaktifkan skrip fallback penulisan aman...")
        for nama_agen in tim_agen.keys():
            sub_dir = os.path.join(root_output_dir, nama_agen.lower())
            os.makedirs(sub_dir, exist_ok=True)
            with open(os.path.join(sub_dir, f"{nama_agen.lower()}_core.py"), "w") as f:
                f.write(f"# Eco-System Component for {nama_agen}\n# Parameter-Driven & AAA Ready.\nprint('{nama_agen} online.')")
            print(f"💾 [FALLBACK SUCCESS] File dasar '{nama_agen.lower()}_core.py' berhasil dibuat.")

    # Pembuatan file config.json eksternal
    config_template = {
        "APP_NAME": nama_proyek,
        "VERSION": "1.0.0",
        "CLOUD_SERVER_URL": "https://api.pabriksoftware.cloud/v1",
        "PAYMENT_GATEWAY_SANDBOX_KEY": "SB-Mid-Xxxxx-Xxxxx",
        "ENCRYPTION_ALGORITHM": "AES-256",
        "LICENSE_CHECK_INTERVAL_DAYS": 1
    }
    with open(os.path.join(root_output_dir, "config.json"), "w") as config_file:
        json.dump(config_template, config_file, indent=4)

    print("\n⚙️ [CONFIG] File 'config.json' berhasil dibuat.")
    print("\n====================================================================")
    print("🚀 [PRODUKSI MASSAL SELESAI] Seluruh komponen fisik berhasil ditulis!")
    print("====================================================================")

# Jalankan Sistem Produksi Baru
jalankan_fase_tiga_bulk(orchestrator_instance=orchestrator)
````
